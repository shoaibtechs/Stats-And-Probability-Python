```python
sizes = np.array([22, 23, 29, 32, 39, 42, 42, 42, 42, 42, 42, 42, 42, 43, 44, 46, 51, 51, 55, 55, 57, 58, 78, 83, 97, 97])
```

```python
sizes.mean()
```

```python
"""ss arry ki average 49 hai....But agar hum array mn dekhen toh ek bhi 49 nhi hai 
...so hum sirf average dekhke teashirts nhi banaiga..."""

"""So iski solution ye hai k hum most occurence value dekhenga is array mn and phir 
us size ki teashirt banainge """
```

```python
from collections import Counter 
cnt = Counter()

for size in sizes:
    cnt[size] += 1
    
cnt.most_common()
```
